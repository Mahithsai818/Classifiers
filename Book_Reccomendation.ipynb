```python
num_ratings_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_ratings_df
```

```python
avg_ratings_df = (
    ratings_with_name
    .groupby('Book-Title')
    .mean(numeric_only=True)
    .reset_index()
)
```

```python
avg_ratings_df.rename(columns = {'Book-Rating': 'avg_rating'},inplace = True)
```

```python
avg_ratings_df
```

```python
popl_df = num_ratings_df.merge(avg_ratings_df, on = 'Book-Title')
popl_df
```

```python
popl_df = popl_df[popl_df['num_rating'] >= 250].sort_values('avg_rating',ascending = False).head(50)
popl_df
```

# Collaborative Filtering Based Recommender System

```python
#lets fetch the users who have rated more than 200 books
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
users_with_200 = x[x].index
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(users_with_200)]
```

```python
#lets fetch books with more than 50 ratings
y = filtered_rating.groupby('Book-Title').count()['Book-Rating'] >= 50
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings
```

```python
pt = final_ratings.pivot_table(index = 'Book-Title',columns = 'User-ID',values = 'Book-Rating')
```

```python
pt
```

```python
pt.fillna(0,inplace = True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores
```

```python
def recommend(book_name):
    index = np.where(pt.index == book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key = lambda x:x[1],reverse = True)[1:6]
    for i in similar_items:
        print(pt.index[i[0]])
```

```python
#lets see some recommendations
```

```python
recommend('1984')
```

```python
recommend('Message in a Bottle')
```

```python
recommend('The Da Vinci Code')
```

```python
recommend('To Kill a Mockingbird')
```
